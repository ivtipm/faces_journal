```python
import cv2  # computer vision
import matplotlib.pyplot as plt
import dlib  # библиотека с алгоритмами МО
from scipy.spatial import distance
```

```python
img = cv2.imread('messi.jpg')

plt.imshow(img)
```

```python
# объект для детекции лиц
detector = dlib.get_frontal_face_detector()

face_rects = detector(img)  # вернёт список из прямоугольников, где обнаружены лица
face_rects
```

```python
frame2 = cv2.rectangle(img, (474, 302), (783, 611), 444, 2)

plt.imshow(img)
```

```python
sp = dlib.shape_predictor('shape_predictor_68_face_landmarks.dat')  # ищет опорные точки на лице
# hape_predictor_68_face_landmarks.dat -- параметры нейросети
# https://github.com/davisking/dlib-models -- другие нейросети
```

```python
face_shape = sp(img, face_rects[0])
face_shape
```

```python
facerec = dlib.face_recognition_model_v1('dlib_face_recognition_resnet_model_v1.dat')
```

```python
face_descriptor = facerec.compute_face_descriptor(img, face_shape)
```

```python
face_descriptor
```
